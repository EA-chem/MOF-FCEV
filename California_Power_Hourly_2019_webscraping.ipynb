```python
import requests, pickle

month_lengths = (31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)
for month in range(1,13):
    for day in range(1,month_lengths[month-1]+1):
        month_str = str(month)
        if len(month_str)==1: 
            month_str = "0" + month_str
        day_str = str(day)
        if len(day_str)==1: 
            day_str = "0" + day_str
        res = requests.get("http://content.caiso.com/green/renewrpt/2019"+month_str+day_str+\
                           "_DailyRenewablesWatch.txt")
        if res.status_code == 404:
            continue
        f = open(("Cali_Data/" + month_str + day_str), "wb")
        pickle.dump(res.text, f)
        f.close()
```

```python
import pickle

# data is list with one list for each hour
data = []
columns = ("Day-Hour", 'GEOTHERMAL', 'BIOMASS', 'BIOGAS', 'SMALL HYDRO', 'WIND TOTAL', 'SOLAR PV', \
           'SOLAR THERMAL', 'NUCLEAR', 'THERMAL', 'IMPORTS', 'HYDRO')

month_lengths = (31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)
for month in range(1,13):
    for day in range(1,month_lengths[month-1]+1):
        month_str = str(month)
        if len(month_str)==1: 
            month_str = "0" + month_str
        day_str = str(day)
        if len(day_str)==1: 
            day_str = "0" + day_str
        
        try:
            f = open(("Cali_Data/"+month_str+day_str), "rb")
            raw = pickle.load(f)
            f.close()
        except:
            continue
        
        lines = raw.split("\n")
        table = []
        for lnum in range(len(lines)):
            table.append([])
            split_line = lines[lnum].split("\t")
            for val in split_line:
                if val not in ["", "\r", "\t"]:
                    table[lnum].append(val)
        for hour in range(1,25):
            hour_str = str(hour)
            if len(hour_str)==1: 
                hour_str = "0" + hour_str
            row = [month_str+day_str+hour_str]
            row.extend(table[1+hour][1:])
            row.extend(table[29+hour][2:])
            data.append(row)

f = open("Cali_Data/00total.pkl", "wb")
pickle.dump(data, f)
f.close()
```
